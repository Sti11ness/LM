```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
class Linear_Regression():
    # Инициализация параметров.
    def __init__(self, learning_rate, no_of_itr):
        self.learning_rate = learning_rate
        self.no_of_itr = no_of_itr
        
    def predict_exp(self, sal):
        # Предсказание опыта работы из зарплаты
        return (sal - self.b) / self.w
    
    def fit(self, X, Y):
        # Количество тренировочных примеров и количество признаков.
        self.m, self.n = X.shape  # Количество строк и столбцов
        # Инициализация веса и смещения
        self.w = np.zeros((self.n, 1))
        self.b = 0
        self.X = X
        self.Y = Y

        # Реализация метода градиентного спуска.
        for i in range(self.no_of_itr):
            self.update_weigths()

    def update_weigths(self):
        Y_prediction = self.predict(self.X)

        # Расчёт градиентов
        dw = -(self.X.T).dot(self.Y - Y_prediction) / self.m
        db = -np.sum(self.Y - Y_prediction) / self.m

        # Обновление весов
        self.w = self.w - self.learning_rate * dw
        self.b = self.b - self.learning_rate * db

    def predict(self, X):
        return X.dot(self.w) + self.b

    def get_weights(self):
        # Возвращаем веса и смещение
        return self.w[0][0], self.b
```

```python
df = pd.read_csv('data_scientist_salaries.csv')
X = df['Exp'].values.reshape(-1, 1)  # Переформатируем для совместимости с моделью
Y = df['Sal'].values.reshape(-1, 1)
```

# Инициилизируем модельку с 2к итерациями и lr 0.0001

```python
# Инициализация и обучение модели
model = Linear_Regression(learning_rate=0.01, no_of_itr=5000)
model.fit(X, Y)
```

```python
# Получение весов для использования в графике
w, b = model.get_weights()
display(w, b)
```
